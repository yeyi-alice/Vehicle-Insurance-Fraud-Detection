```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel('data_new.xlsx')
df
```

```python
X = df[['AccidentArea',
        'Sex',
        'MaritalStatus',
        'Age', 
        'Fault', 
        'VehicleCategory', 
        'VehiclePrice', 
        'RepNumber', 
        'Deductible', 
        'DriverRating', 
        'Days_Policy_Accident',
        'Days_Policy_Claim',
        'PastNumberOfClaims',
        'AgeOfVehicle',
        'PoliceReportFiled',
        'WitnessPresent',
        'AgentType',
        'NumberOfSuppliments',
        'AddressChange_Claim',
        'NumberOfCars',
        'BasePolicy',
        'Date_Diff']]

    

y = df['FraudFound_P']

print(X.shape)
print(y.shape)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder


categorical_columns =  ['AccidentArea',
                        'Sex',
                        'MaritalStatus',
                        'Fault', 
                        'VehicleCategory', 
                        'VehiclePrice',
                        'Days_Policy_Accident',
                        'Days_Policy_Claim',
                        'PastNumberOfClaims',
                        'AgeOfVehicle',
                        'PoliceReportFiled',
                        'WitnessPresent',
                        'AgentType',
                        'NumberOfSuppliments',
                        'AddressChange_Claim',
                        'NumberOfCars',
                        'BasePolicy']


preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), categorical_columns)
    ],
    remainder='passthrough'  # This will keep non-categorical columns unchanged
)

# Step 4: Fit and transform X_train
X_train_encoded = preprocessor.fit_transform(X_train)

# Step 5: Transform X_test using the same transformer
X_test_encoded = preprocessor.transform(X_test)

print(X_train_encoded.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

clf = RandomForestClassifier(random_state=0)
clf.fit(X_train_encoded, y_train)
y_pred_train = clf.predict(X_train_encoded)
y_pred = clf.predict(X_test_encoded)
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy of the Random Forest Classifier:", accuracy)
```

```python
from sklearn.model_selection import RandomizedSearchCV
clf2 = RandomForestClassifier(random_state=0)

param_dist = {
    "n_estimators": list(range(1, 1000)),
    "max_depth": list(range(1, 10)),
    "min_samples_split": list(range(5, 100))
}


random_search = RandomizedSearchCV(clf2, param_distributions=param_dist, n_iter=1000, cv=10, random_state=0, n_jobs=-1, scoring='recall')
random_search.fit(X_train_encoded, y_train)


best_params = random_search.best_params_
best_score = random_search.best_score_

print("Best Hyperparameters:", best_params)
print("Mean Cross-Validated Score of the Best Estimator:", best_score)

best_estimator = random_search.best_estimator_

y_pred2_train = best_estimator.predict(X_train_encoded)
y_pred2 = best_estimator.predict(X_test_encoded)
accuracy2 = accuracy_score(y_test, y_pred2)

print("Accuracy of the Random Forest Classifier (Tuned):", accuracy2)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, confusion_matrix
accuracy_train_combined = accuracy_score(y_train, y_pred_train)
precision_train_combined = precision_score(y_train, y_pred_train)
recall_train_combined = recall_score(y_train, y_pred_train)
roc_auc_train_combined = roc_auc_score(y_train, clf.predict_proba(X_train_encoded)[:, 1])
confusion_matrix_train_combined = confusion_matrix(y_train, y_pred_train)

accuracy_test_combined = accuracy_score(y_test, y_pred)
precision_test_combined = precision_score(y_test, y_pred)
recall_test_combined = recall_score(y_test, y_pred)
roc_auc_test_combined = roc_auc_score(y_test, clf.predict_proba(X_test_encoded)[:, 1])
confusion_matrix_test_combined = confusion_matrix(y_test, y_pred)

print("Training Set Metrics (Not Tuned):")
print("Accuracy:", accuracy_train_combined)
print("Precision:", precision_train_combined)
print("Recall:", recall_train_combined)
print("ROC AUC:", roc_auc_train_combined)
print("Confusion Matrix:\n", confusion_matrix_train_combined)

print("\nTest Set Metrics (Not Tuned):")
print("Accuracy:", accuracy_test_combined)
print("Precision:", precision_test_combined)
print("Recall:", recall_test_combined)
print("ROC AUC:", roc_auc_test_combined)
print("Confusion Matrix:\n", confusion_matrix_test_combined)



```
